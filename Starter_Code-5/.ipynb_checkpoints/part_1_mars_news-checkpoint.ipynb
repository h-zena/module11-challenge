````python
#creating Beautiful Soup object
from bs4 import BeautifulSoup
import requests

url = 'https://static.bc-edx.com/data/web/mars_news/index.html'

# send a GET request to the URL
response = requests.get(url)

# analyze the HTML content of the webpage
soup = BeautifulSoup(response.content, 'html.parser')

# extracted text elements from the webpage
text_elements = soup.find_all(text=True)

# print the extracted text elements
for element in text_elements:
    print(element)
````

### Step 3: Store the Results

Extract the titles and preview text of the news articles that you scraped. Store the scraping results in Python data structures as follows:

* Store each title-and-preview pair in a Python dictionary. And, give each dictionary two keys: `title` and `preview`. An example is the following:

  ```python
  {'title': "NASA's MAVEN Observes Martian Light Show Caused by Major Solar Storm", 
   'preview': "For the first time in its eight years orbiting Mars, NASA’s MAVEN mission witnessed two different types of ultraviolet aurorae simultaneously, the result of solar storms that began on Aug. 27."
  }
  ```

* Store all the dictionaries in a Python list.

* Print the list in your notebook.

````python
# Create an empty list to store the dictionaries

url = 'https://static.bc-edx.com/data/web/mars_news/index.html'

response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

articles = []

# find anything containing the titles and preview text of the news articles
article_elements = soup.find_all('div', class_='list_text')

# looped through the article elements and extracted the title and preview text
for article in article_elements:
    title = article.find('div', class_='content_title').text.strip()
    preview = article.find('div', class_='article_teaser_body').text.strip()
    article_dict = {'title': title, 'preview': preview}
    articles.append(article_dict)

# Print the list of dictionaries
for article in articles:
    print(article)
````

````python
# Loop through the text elements
# Extract the title and preview text from the elements
# Store each title and preview pair in a dictionary
# Add the dictionary to the list

url = 'https://static.bc-edx.com/data/web/mars_news/index.html'

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
articles = []

# Loop through the text elements
for text_element in soup.find_all(text=True):
    # Extract the title and preview text from the elements
    if text_element.parent.name == 'div' and text_element.parent.get('class') == ['list_text']:
        title_element = text_element.parent.find('div', class_='content_title')
        preview_element = text_element.parent.find('div', class_='article_teaser_body')
        
     # Store each title and preview pair in a dictionary
        if title_element and preview_element:
            title = title_element.get_text().strip()
            preview = preview_element.get_text().strip()
            article_dict = {'title': title, 'preview': preview}
            
            # Add the dictionary to the list
            articles.append(article_dict)

# Print the list of dictionaries
for article in articles:
    print(article)
````

````python
# Print the list to confirm success

print(articles)
````
